```python
import pandas as pd
```

```python
# 전체 곡 파일 불러오기
song_eomi = pd.read_csv('song_okt_eomi_df.csv')
```

```python
# 여자아이돌 곡 파일 불러오기
female_eomi = pd.read_csv('female_okt_eomi_df.csv')
```

```python
# 남자아이돌 곡 파일 불러오기
male_eomi = pd.read_csv('male_okt_eomi_df.csv')
```

```python
# 중복값을 지우고 유니크한 값만 보는 df.
# (중복값 지우지 않은 df도 살려놓는 것이 좋을 듯)
song_e_unique = song_eomi.drop_duplicates(subset='word')
female_e_unique = female_eomi.drop_duplicates(subset='word')
male_e_unique = male_eomi.drop_duplicates(subset='word')
```

```python
female_e_unique = female_e_unique.rename(index=str, columns={"count": "female"})
```

```python
female_e_unique.head()
```

```python
male_e_unique = male_e_unique.rename(index=str, columns={"count": "male"})
```

```python
male_e_unique.head()
```

```python
fm_merge = pd.merge(female_e_unique, male_e_unique, on=['word','parts_of_speech'], how='outer')
fm_merge.head()
```

```python
song_eomi['parts_of_speech'].unique()
```

```python
# 유니크한 값만 count로 sort해서 보기(품사별로)
fm_merge_V = fm_merge[fm_merge['parts_of_speech'] == 'Verb'].sort_values(by=['female'], ascending = False)
fm_merge_A = fm_merge[fm_merge['parts_of_speech'] == 'Adverb'].sort_values(by=['female'], ascending = False)
fm_merge_Adj = fm_merge[fm_merge['parts_of_speech'] == 'Adjective'].sort_values(by=['female'], ascending = False)
fm_merge_Alpha = fm_merge[fm_merge['parts_of_speech'] == 'Alpha'].sort_values(by=['female'], ascending = False)
fm_merge_Eomi = fm_merge[fm_merge['parts_of_speech'] == 'Eomi'].sort_values(by=['female'], ascending = False)
```
